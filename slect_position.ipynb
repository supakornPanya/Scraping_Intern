```python
import csv
import requests
import pandas as pd
import time
import random
import json
import os
from dotenv import load_dotenv

with open('listNamePosition.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    listNamePosition = next(reader)
```

```python
print(listNamePosition)
```

```python
# URL พร้อม Query Parameters (เว้น page ไว้ใส่ค่า)
API_URL_TEMPLATE = "https://cedtintern.cp.eng.chula.ac.th/api/sessions/5/openings/{}/bookmark"

cookie_value = os.getenv("COOKIE")

HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36",
    "Accept": "application/json, text/plain, */*",
    "Content-Type": "application/json",
    "Origin": "https://cedtintern.cp.eng.chula.ac.th",
    "Referer": "https://cedtintern.cp.eng.chula.ac.th/opening?page=1",
    "Cookie": cookie_value
}

success_count = 0
error_count = 0

for id in listNamePosition:
    url = API_URL_TEMPLATE.format(id)
    
    try:
        print(f"Bookmarking {id}...", end=" ")
        # IMPORTANT: Use POST method (not PUT) with empty body
        response = requests.post(url, 
                                 headers=HEADERS, 
                                 timeout=10,
                                 json={}  # Empty payload
                            )
        
        if response.status_code == 200:
            print(f"[OK!]")
            success_count += 1
        else:
            print(f"[ERR] Status {response.status_code}: {response.text}")
            error_count += 1

    except Exception as e:
        print(f"[EXCEPTION] {e}")
        error_count += 1

    # Random delay to avoid being blocked
    delay = random.uniform(2.0, 5.0)
    time.sleep(delay)
    
print(f"\n✓ Completed: {success_count} success, {error_count} errors")
```
